```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from sklearn.preprocessing import Normalizer
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv('files_for_lab/csv_files/marketing_customer_analysis.csv')
data
```

```python
# I am applying some of the statements of the function created in lab-customer-analysis-round-2 to deal with the data
# I will not explore the data again since it has already been done in lab-customer-analysis-round-3 and it was mostly plotting

def data_cleaning(data):
    data.columns = [columns_name.lower() for columns_name in data.columns]
    data.columns = [columns_name.replace(' ', '_') for columns_name in data.columns]
    data.dropna(inplace = True)
    data['effective_to_date'] = pd.to_datetime(data['effective_to_date'], errors='coerce')
    data['month'] = data['effective_to_date'].dt.month
    return data
```
